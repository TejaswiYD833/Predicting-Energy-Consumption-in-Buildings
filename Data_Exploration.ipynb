```python
def details(df):
    print('Column Types')
    print(df.dtypes)
    print('------------------')
    print('Missing Values')
    print(df.isna().sum())
    print('------------------')
    print('Description')
    print(df.describe().applymap('{:,.2f}'.format))

```
